## Arquitectura
- `source_inegi/`: insumo local inmutable
- `src/pipeline/`: logica reusable
- `data/intermediate/`: trazabilidad (manifest, discovery)
- `outputs/`: tablas finales para integracion posterior

```python
from pathlib import Path
root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd()
print('Root:', root)
for rel in ['src/pipeline', 'data/intermediate', 'outputs', 'docs']:
    p = root / rel
    print(rel, 'OK' if p.exists() else 'MISSING')
```
